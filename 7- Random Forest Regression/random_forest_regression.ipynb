```python
# Random Forest Regression

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importar el dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Ajustar la regresión con el dataset
from sklearn.ensemble import RandomForestRegressor

regression = RandomForestRegressor(n_estimators=300, random_state=0)
```

```python
y_pred = regression.fit(x, y)

# Predicción de nuestros modelos con Random Forest
y_pred = regression.predict([[6.5]])
y_pred
```

```python
# Visualización de los resultados del RFR
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regression.predict(x_grid), color = 'blue')
plt.title('Modelo de Regresión RFR')
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo (en $)')
plt.show()
```
