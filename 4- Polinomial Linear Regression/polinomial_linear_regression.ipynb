```python
# Regresion polinomial
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Importar el data set
dataset = pd.read_csv("Position_Salaries.csv")
```

```python
# variables independientes X
# [:, :-1] -> todas las filas, todas las columnas menos la ultima
x = dataset.iloc[:, 1:2].values

# variables dependientes Y
y = dataset.iloc[:, 2].values  # [:, 2] -> todas las filas, columna 2
```

```python
# Ajustar la regresion lineal con el dataset
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(x, y)

regression.score(x, y)
```

```python
# Ajustar la regresion polinomica con el dataset

poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)

# Ajustar la regresion lineal con el dataset

lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
# Visualizacion de los resultados del modelo lineal
import matplotlib.pyplot as plt

plt.scatter(x, y, color="red")

plt.plot(x, regression.predict(x), color="blue")

plt.title("Modelo de regresion lineal")

plt.xlabel("Posicion del empleado")

plt.ylabel("Sueldo (en $)")
```

```python
# Visualizacion de los resultados del modelo polinomico
plt.scatter(x, y, color="red")

plt.plot(x, lin_reg_2.predict(x_poly), color="blue")

plt.title("Modelo de regresion polinomial")

plt.xlabel("Posicion del empleado")

plt.ylabel("Sueldo (en $)")
```

```python
import numpy as np
# Ajustar la regresion polinomica con el dataset

poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)

# Ajustar la regresion lineal con el dataset

lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)

# Visualizacion de los resultados del modelo polinomico
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))

plt.scatter(x, y, color="red")

plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color="blue")

plt.title("Modelo de regresion polinomial")

plt.xlabel("Posicion del empleado")

plt.ylabel("Sueldo (en $)")
```

```python
# Prediccion de nuestros modelos
regression.predict([[6.5]])
```

```python
# Prediccion de nuestros modelos
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
