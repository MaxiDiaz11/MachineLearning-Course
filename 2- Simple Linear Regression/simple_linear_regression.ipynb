```python
# Regresion lineal simple
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Importar el data set
dataset = pd.read_csv("Salary_Data.csv")
```

```python
# variables independientes X
# [:, :-1] -> todas las filas, todas las columnas menos la ultima
x = dataset.iloc[:, :-1].values

# variables dependientes Y
y = dataset.iloc[:, 1].values  # [:, 1] -> todas las filas, columna 1
```

```python
# Dividir el data set en conjunto de entrenamiento y conjunto de testing
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=1 / 3, random_state=0
)
```

```python
# Crear modelo de regresion lineal simple con el conjunto de entrenamiento
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

# Entrenar el modelo con el conjunto de entrenamiento
regression.fit(x_train, y_train)
```

```python
# Predecir el conjunto de test
y_pred = regression.predict(x_test)
```

```python
# Visualizar los resultados de entrenamiento
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, color="red")

# Grafica de la recta de regresion
plt.plot(x_train, regression.predict(x_train), color="blue")

# Titulo
plt.title("Sueldo vs Años de experiencia (Conjunto de entrenamiento)")

# Etiqueta eje x
plt.xlabel("Años de experiencia")

# Etiqueta eje y
plt.ylabel("Sueldo")
```
