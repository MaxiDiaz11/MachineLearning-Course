```python
# SVR

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importar el dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
from sklearn.svm import SVR

regression = SVR(kernel = 'rbf')

regression.fit(x, y)
```

```python
pred = np.array([6.5]).reshape(1, 1)
y_pred = regression.predict(pred)
```

```python
# Visualización de los resultados del SVR
plt.scatter(x, y, color = 'red')
plt.plot(x, regression.predict(x), color = 'blue')
plt.title('Modelo de Regresión SVR')
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo (en $)')
plt.show()
```

```python
# Escalado de variables
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()

x_sc = sc_x.fit_transform(x)
y_sc = sc_y.fit_transform(y.reshape(-1, 1))
```

```python
regression = SVR(kernel = 'rbf')

regression.fit(x_sc, y_sc)
```

```python
pred = np.array(sc_x.transform([[6.5]]))
y_pred = regression.predict(pred).reshape(-1, 1)
y_pred = sc_y.inverse_transform(y_pred)
y_pred
```

```python
# Visualización de los resultados del SVR
x_grid = np.arange(min(x_sc), max(x_sc), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x_sc, y_sc, color = 'red')
plt.plot(x_grid, regression.predict(x_grid), color = 'blue')
plt.title('Modelo de Regresión SVR')
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo (en $)')
plt.show()
```
